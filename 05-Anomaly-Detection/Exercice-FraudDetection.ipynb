# I. Imbalanced data and metrics

## I.1. Imbalanced data and the accuracy paradox

Skewed data is highly imbalanced data: meaning most of the samples are from one class.

Most of the time, when having a higly imbalanced dataset, the accuracy is not a good evaluation metrics. Let's see why.

Let's take the example of a rare disease, that occurs only for 1 over 100 patient test. So you have a set of features to predict whether a patient is infected or not.

Then this is a basic binary classification problem, with the following data:
<ul>
<li> class 0: the patient is not infected (99% of the whole dataset </li>
<li> class 1: the patient is infected (1% of the whole dataset) </li>
</ul>

What if, for some reason, our machine learning model predicts always class 0? What would be our accuracy?

The accuracy would be 99%...

That's what we call the accuracy paradox: with highly imbalanced dataset, the accuracy is not a good evaluation metrics.

Then what can we use? What other metrics do we know?

## I.2. Performance measurement with imbalanced dataset

Well, we also know the following metrics:
<ul>
<li> Precision </li>
<li> Recall </li>
<li> F1-score </li>
<li> ROC </li>
<li> ROC AUC </li>
</ul>

Let's see how it would work on our examples with those metrics if our model almost always predicts class 0

|   | **Not infected (class 0)** | **Infected (class 1)** | 
|------|------|------|
| **Predicted not infected** 👍 | 979 | 9 |
| **Predicted infected** 👎 | 11 | 1 |

What are the TP, TN, FP, FN values out of this?

* True Positive (TP): 1
* True Negative (TN): 979
* False Positive (FP): 11
* False Negative (FN): 9

Now let's compute our metrics.

**Accuracy:**

First as a reminded, let's compute the accuracy on this example:
$$Accuracy = \frac{TP + TN}{TP + TN + FN + FP}$$

So here Accuracy = 980/1000 = 98 %

Quite a good accuracy, right? What about other metrics now?

**Precision**

The precision formula is:
$$Precision = \frac{TP}{TP + FP}$$

So here Precision = 1/12 = 8.3%! 

Wow, this is not good compared to accuracy?

**Recall**

The recall formula is:
$$Recall = \frac{TP}{P} = \frac{TP}{TP + FN}$$

So here Recall = 1/10 = 10%!

Not good either...

**F1-score**

The F1-score formula is the following:
$$F1\_score = \frac{2 * (Precision * Recall)}{Precision + Recall}$$

F1-score = 9.1 %

So what do you think about this model? Even with an accuracy of 98%, would you use it in real life?

Just as a reminder, you can compute all those metrics using scikit-learn, here is how:

```python
# import libraries
import numpy as np
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score

# recreate the dataset above
y_true = np.zeros(1000)
y_true[-10:]=1
y_pred = np.zeros(1000)
y_pred[:11]=1
y_pred[-1:] = 1

print('accuracy:', accuracy_score(y_true, y_pred))
print('precision:', precision_score(y_true, y_pred))
print('recall:', recall_score(y_true, y_pred))
print('F1-score:', f1_score(y_true, y_pred))
```

Quite easy, right? Now your turn to work on the credit card fraud detection challenge!

# II. Approaches to handle imbalanced data

## II.1. Undersampling

One possible approach is to undersample your initial dataset. Let's look again at our example.

| **Not infected (class 0)** | **Infected (class 1)** | 
|:----:|:----:|
 | 990 | 10 |

So we have only $\frac{10}{1000} = 1\%$ of infected samples.

To undersample, we would keep all our observations from underrepresented class (here *infected*) and keep only a subsample of the other class.

For example using Random Undersampling, we would keep our 10 infected samples, and keep only 90 randomly chosen not infected samples. So at the end we would have the following dataset:

| **Not infected (class 0)** | **Infected (class 1)** | 
|:----:|:----:|
 | 90 | 10 |
 
So we would have now $\frac{10}{100} = 10\%$ of infected samples. So our dataset is much less imbalanced.

But the drawback is that we lose many potentially useful samples, and we may not keep the original distribution of the data.

This method can be applied using the library called *Imbalanced Learn* (can be installed using `conda install imbalanced-learn` if not already done).

In particular, you can use the following class:

`class imblearn.under_sampling.RandomUnderSampler(sampling_strategy='auto', return_indices=False, random_state=None, replacement=False, ratio=None)`

You can find an example and more details about undersampling [here](https://imbalanced-learn.org/stable/).

## II.2. Random oversampling: SMOTE

Another approach would be to do the opposite: instead of removing *not infected* samples, we would artificially **add** *infected* samples.

The most naive way of doing this would be to duplicate under represented labels. For example in our example you would duplicate the 10 *infected* samples 10 times, thus having 100 *infected* labels. This technique has drawbacks, it does not create new information and may lead to overfitting on the duplicated data.

A smart way of oversampling is the so called SMOTE (for Synthetic Minority Oversampling Technique). 

The idea is to choose wisely new samples using a k-NN on original data. To do so, a new sample $x_{new}$ will be computed based on the k nearest neighboors of an already existing sample $x_i$ with the following formula:

$$
x_{new} = x_i + \lambda \times (x_{zi} - x_i)
$$

Where $x_{zi}$ is one of the k nearest neighboors of $x_i$ and $\lambda$ is a random value between 0 and 1.

`class imblearn.over_sampling.SMOTE(sampling_strategy='auto', random_state=None, k_neighbors=5, m_neighbors='deprecated', out_step='deprecated', kind='deprecated', svm_estimator='deprecated', n_jobs=1, ratio=None)`
